```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("forLoB/logLoB.csv", sep=';', names=['nx', 'np', 'type', 'time'], skiprows=1)
save = df


for VAR in ['OMP','OMPTask','TBB','TBBRange']:
    var = df.loc[df["type"] == VAR]
    # var = var.drop('type', 1)

    cols = var["np"].values.ravel()
    nb_threads_ite = pd.unique(cols)
    nb_threads_ite.sort()

    gb = var.groupby(['np', 'nx']).agg({
        'time' : np.min,
    })

    gb = gb.reset_index()
    gb.sort_values(by=['np'])

    plt.figure(VAR)
    
    for nt in nb_threads_ite :
        df = gb[gb['np'] == nt]
        
        s = df['nx'].to_numpy()
        t = df['time'].to_numpy()
        
        plt.plot(s, t, label="Nb thread = {}".format(nt))
    
    plt.legend()
    plt.xlabel("Flock size (number of Agents)")
    plt.ylabel("Time (seconds)")
    plt.grid(visible=True, axis='both')
    plt.title(VAR)
    
    df = save

plt.show()
```
