```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 1 Dense MV
## Plot MPI logs

```python
df = pd.read_csv("./densemv_MPI_logs.csv", sep=';', names=['nx', 'np', 'num_p', 'time'], skiprows=1)

# cols = df["np"].values.ravel()[1:]
# nps = pd.unique(cols)

cols = df["nx"].values.ravel()
nxs = pd.unique(cols)

# print(df)
gb = df.groupby(['nx', 'np']).agg({
    'time' : np.min,
})


gb = gb.reset_index()

nxs.sort()
for nx in nxs:
    pack = gb[gb['nx'] == nx]
    t = pack['time'].to_numpy()
    nb_p = pack['np'].to_numpy()
    
    
    plt.plot(nb_p, t, label="nx = {}".format(nx))
    
plt.legend()
plt.xlabel("Nb proc")
plt.ylabel("Time")
plt.grid(visible=True, axis='both')
plt.show()
   
```

## Sequential

```python
df = pd.read_csv("./densemv_seq_logs.csv", sep=';', names=['nx', 'np', 'time'], skiprows=1)
gb = df.groupby(['nx', 'np']).agg({
    'time' : np.min,
})
gb = gb.reset_index()

cols = df["nx"].values.ravel()
nxs = pd.unique(cols)
nxs.sort

print(nxs)

for nx in nxs:
    pack = gb[gb['nx'] == nx]
    t = pack['time'].to_numpy()
    nb_p = pack['np'].to_numpy()
    
    
    plt.plot(nb_p, t, label="nx = {}".format(nx))
    
plt.legend()
plt.xlabel("Nb proc")
plt.ylabel("Time")
plt.grid(visible=True, axis='both')
plt.show()
   
```

## Final plot Dense

```python
df = pd.read_csv("./densemv_MPI_logs.csv", sep=';', names=['nx', 'np', 'num_p', 'time'], skiprows=1)
# print(df)
gb = df.groupby(['nx', 'np']).agg({
    'time' : np.max,
})
gb = gb.reset_index()

df2 = pd.read_csv("./densemv_seq_logs.csv", sep=';', names=['nx', 'np', 'time'], skiprows=1)
gb2 = df2.groupby(['nx', 'np']).agg({
    'time' : np.max,
})
gb2 = gb2.reset_index()
# cols = df["np"].values.ravel()[1:]
# nps = pd.unique(cols)

cols = df["nx"].values.ravel()
nxs = pd.unique(cols)

colors = [ 'xb', 'xr', 'xg', 'xm']

nxs.sort()
for nx, c in zip(nxs, colors):
    pack = gb[gb['nx'] == nx]
    t = pack['time'].to_numpy()
    nb_p = pack['np'].to_numpy()
    
    pack2 = gb2[gb2['nx'] == nx]
    t2 = pack2['time'].to_numpy()
    nb_p2 = pack2['np'].to_numpy()

    plt.plot(nb_p, t, c+"-", label="nx = {}".format(nx))
    plt.plot(nb_p2, t2, c+'--')

plt.plot(0,0, 'k--', label="seq")
plt.legend()

plt.xlabel("Nb proc")
plt.ylabel("Time")
plt.grid(visible=True, axis='both')
plt.show()
   
```
