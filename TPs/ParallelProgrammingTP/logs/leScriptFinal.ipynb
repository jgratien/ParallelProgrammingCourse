```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("MPI_logs_Dense.csv", sep=';', names=['nx', 'np', 'num_p', 'time'], skiprows=1)
dense = df.groupby(['nx', 'np']).agg({
    'time' : np.max,
})
dense = dense.reset_index()

df = pd.read_csv("seq_logs_Dense.csv", sep=';', names=['nx', 'np', 'time'], skiprows=1)
denseSeq = df.groupby(['nx', 'np']).agg({
    'time' : np.max,
})
denseSeq = denseSeq.reset_index()

df = pd.read_csv("MPI_logs_Sparse.csv", sep=';', names=['nx', 'np', 'num_p', 'time'], skiprows=1)
sparse = df.groupby(['nx', 'np']).agg({
    'time' : np.max,
})
sparse = sparse.reset_index()

df = pd.read_csv("seq_logs_Sparse.csv", sep=';', names=['nx', 'np', 'time'], skiprows=1)
sparseSeq = df.groupby(['nx', 'np']).agg({
    'time' : np.max,
})
sparseSeq = sparseSeq.reset_index()
```

```python
dense50 = dense.loc[dense['nx']==50]
dense100 = dense.loc[dense['nx']==100]
dense150 = dense.loc[dense['nx']==150]
dense200 = dense.loc[dense['nx']==200]

denseSeq50 = denseSeq.loc[denseSeq['nx']==50]
denseSeq100 = denseSeq.loc[denseSeq['nx']==100]
denseSeq150 = denseSeq.loc[denseSeq['nx']==150]
denseSeq200 = denseSeq.loc[denseSeq['nx']==200]

sparse50 = sparse.loc[sparse['nx']==50]
sparse100 = sparse.loc[sparse['nx']==100]
sparse150 = sparse.loc[sparse['nx']==150]
sparse200 = sparse.loc[sparse['nx']==200]

sparseSeq50 = sparseSeq.loc[sparseSeq['nx']==50]
sparseSeq100 = sparseSeq.loc[sparseSeq['nx']==100]
sparseSeq150 = sparseSeq.loc[sparseSeq['nx']==150]
sparseSeq200 = sparseSeq.loc[sparseSeq['nx']==200]
```

```python
plt.grid()
plt.xlabel("Processors (number)")
plt.ylabel("Time (seconds)")
plt.title("Performance : Dense")
plt.plot(dense50.np,dense50.time,label="mpi np = 50",color="r")
plt.plot(dense100.np,dense100.time,label="mpi np = 100",color="b")
plt.plot(dense150.np,dense150.time,label="mpi np = 150",color="g")
plt.plot(dense200.np,dense200.time,label="mpi np = 200",color="m")
plt.plot(denseSeq50.np,denseSeq50.time,label="seq np = 50",color="r",linestyle='dotted')
plt.plot(denseSeq100.np,denseSeq100.time,label="seq np = 100",color="b",linestyle='dotted')
plt.plot(denseSeq150.np,denseSeq150.time,label="seq np = 150",color="g",linestyle='dotted')
plt.plot(denseSeq200.np,denseSeq200.time,label="seq np = 200",color="m",linestyle='dotted')
plt.legend()
```

```python
plt.grid()
plt.xlabel("Processors (number)")
plt.ylabel("Time (seconds)")
plt.title("Performance : Dense (Zoom)")
plt.plot(dense50.np,dense50.time,label="mpi np = 50",color="r")
plt.plot(dense100.np,dense100.time,label="mpi np = 100",color="b")
plt.plot(dense150.np,dense150.time,label="mpi np = 150",color="g")
# plt.plot(dense200.np,dense200.time,label="mpi np = 200",color="m")
plt.plot(denseSeq50.np,denseSeq50.time,label="seq np = 50",color="r",linestyle='dotted')
plt.plot(denseSeq100.np,denseSeq100.time,label="seq np = 100",color="b",linestyle='dotted')
plt.plot(denseSeq150.np,denseSeq150.time,label="seq np = 150",color="g",linestyle='dotted')
# plt.plot(denseSeq200.np,denseSeq200.time,label="seq np = 200",color="m",linestyle='dotted')
plt.legend()
```

```python
cinquante = np.array(dense50.time)
seqCinquante = np.array(denseSeq50.time)
accelCinquante = seqCinquante / cinquante
accelCinquante
cent = np.array(dense100.time)
seqCent = np.array(denseSeq100.time)
accelCent = seqCent / cent
accelCent
centCinquante = np.array(dense150.time)
seqCentCinquante = np.array(denseSeq150.time)
accelCentCinquante = seqCentCinquante / centCinquante
accelCentCinquante
deuxCent = np.array(dense200.time)
seqDeuxCent = np.array(denseSeq200.time)
accelDeuxCent = seqDeuxCent / deuxCent
accelDeuxCent
```

```python
plt.grid()
plt.xlabel("Processors (number)")
plt.ylabel("Acceleration ('%' of time)")
plt.title("Acceleration : Dense")
plt.plot(dense50.np,accelCinquante,label="mpi np = 50",color="r")
plt.plot(dense100.np,accelCent,label="mpi np = 100",color="b")
plt.plot(dense150.np,accelCentCinquante,label="mpi np = 150",color="g")
plt.plot(dense200.np,accelDeuxCent,label="mpi np = 200",color="m")
plt.legend()
```

```python
plt.grid()
plt.xlabel("Processors (number)")
plt.ylabel("Time (seconds)")
plt.title("Performance : Sparse")
plt.plot(sparse50.np,sparse50.time,label="mpi np = 50",color="r")
plt.plot(sparse100.np,sparse100.time,label="mpi np = 100",color="b")
plt.plot(sparse150.np,sparse150.time,label="mpi np = 150",color="g")
plt.plot(sparse200.np,sparse200.time,label="mpi np = 200",color="m")
plt.plot(sparseSeq50.np,sparseSeq50.time,label="seq np = 50",color="r",linestyle='dotted')
plt.plot(sparseSeq100.np,sparseSeq100.time,label="seq np = 100",color="b",linestyle='dotted')
plt.plot(sparseSeq150.np,sparseSeq150.time,label="seq np = 150",color="g",linestyle='dotted')
plt.plot(sparseSeq200.np,sparseSeq200.time,label="seq np = 200",color="m",linestyle='dotted')
plt.legend()
```

```python
plt.grid()
plt.xlabel("Processors (number)")
plt.ylabel("Time (seconds)")
plt.title("Performance : Sparse (Zoom)")
plt.plot(sparse50.np,sparse50.time,label="mpi np = 50",color="r")
plt.plot(sparse100.np,sparse100.time,label="mpi np = 100",color="b")
plt.plot(sparse150.np,sparse150.time,label="mpi np = 150",color="g")
# plt.plot(sparse200.np,sparse200.time,label="mpi np = 200",color="m")
plt.plot(sparseSeq50.np,sparseSeq50.time,label="seq np = 50",color="r",linestyle='dotted')
plt.plot(sparseSeq100.np,sparseSeq100.time,label="seq np = 100",color="b",linestyle='dotted')
plt.plot(sparseSeq150.np,sparseSeq150.time,label="seq np = 150",color="g",linestyle='dotted')
# plt.plot(sparseSeq200.np,sparseSeq200.time,label="seq np = 200",color="m",linestyle='dotted')
plt.legend()
```
